```julia
using Printf, DelimitedFiles, Plots 
```

```julia
function readFECvS(bondD, β; Jz = 1.0, Jxy=1.0, wid = 1)
    Path = @sprintf "/data/sliang/JuliaCMPO/XXZ_2D_helical/Jz_%.2f_Jxy_%.2f_wid_%02i/bondD_%02i_CMPS/CheckPoint_beta_%.2f/Obsv_FECvS.txt" Jz Jxy wid bondD β
    return readdlm(Path, skipstart = 2)
end
```

```julia
dtrg = readdlm("/data/sliang/CMPO/XXZ/Jxy_1.00_Jz_1.00/DTRG_Result/Obsv_F_bondD_256.txt", skipstart=14)[1:5,:]
```

```julia
β = 16.0; W = 3
d08 = readFECvS(8,β,wid = W);
d10 = readFECvS(10,β,wid = W);
d12 = readFECvS(12,β,wid = W);
```

```julia
plot(d08[:,1], d08[:,2], line=(2), label="D=8")
plot!(d10[:,1], d10[:,2], line=(2), label="D=10")
plot!(d12[:,1], d12[:,2], line=(2), label="D=12")
plot!(xlabel="step", ylabel="F", title=@sprintf "width=%i, β=%.2f" W β)
```

```julia
plot(d08[:,1], d08[:,3], line=(2), label="D=8")
plot!(d10[:,1], d10[:,3], line=(2), label="D=10")
plot!(d12[:,1], d12[:,3], line=(2), label="D=12")
plot!(xlabel="step", ylabel="E", title=@sprintf "width=%i, β=%.2f" W β)
```
